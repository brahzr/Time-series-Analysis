**Def**: A time series is a set of data points ordered in time.
The data is equally spaced in time, meaning that it was recorded at every hour, minute, month,
or quarter. Typical examples of time series include the closing value of a stock, a household’s
electricity consumption, or the temperature outside. <br>

**xt**

**t**: A prior or lag time <br>
**t-n**: A current time and point of reference <br>
**t+n**: A future or forecast time <br>

**Time resampling and Frequency**

```python
#get max values from meta yearly
df_meta.resample(rule = "A").max()
```

```python
#get monthly minumum 
df_meta.resample(rule = "m").min()
```

```python
#get monthly average of data
df_meta.resample(rule = "m").mean()
```

```python
#get quarterly mean 
df_meta.resample(rule = "Q").mean()
```

```python
#BQS get Business quarterly
df_meta.resample(rule= "BQS").max()
```

```python
# get maximum values of High in a monthly frequency
df_meta["High"].resample(rule = "m").max()
```

```python
#let's plot that
df_gold["High"].resample(rule = "m").max().plot(kind = "line", figsize = (6,4))
#line, box, pie, area, hist
```
